```python
# Calculate sample mean
sample_mean = np.mean(data)
```

```python
# Calculate variance and standard deviation
variance = np.var(data)
std_dev = np.sqrt(variance)
```

```python
# Calculate moments about the mean
mu1 = 0
mu3 = np.mean(((data - sample_mean) / std_dev) ** 3)
mu4 = np.mean(((data - sample_mean) / std_dev) ** 4)
```

```python
# Calculate Pearson skewness and kurtosis coefficients
beta1 = mu3 / (std_dev ** 3)
beta2 = mu4 / (std_dev ** 4)
```

```python
# Classify dataset based on kurtosis
if beta2 > 3:
    classification = 'Leptokurtic'
elif beta2 < 3:
    classification = 'Platykurtic'
else:
    classification = 'Mesokurtic'
```
